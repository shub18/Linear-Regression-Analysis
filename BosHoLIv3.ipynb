```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston 

boston1 = load_boston()


print(boston1)

```

```python
#Create a DataFrame

df1 = pd.DataFrame(boston1.data, columns=['CRIM' ,'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO',
 'B', 'LSTAT'])

print(df1)
```

```python
df1['MEDV'] = boston1.target

print(df1)
```

```python
print(df1.describe())
```

```python
corr_matric = df1.corr()
plt.figure(figsize=(18, 10))
sns.heatmap(corr_matric, annot=True, cmap='Blues')
plt.show()
```

```python
print(df1.describe())
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

mm = StandardScaler()

mm1 = mm.fit_transform(df1)

print(mm1)
```

```python
#train test sample creation 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(mm1, boston1.target, train_size=0.85)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Model fitting

from sklearn.linear_model import LinearRegression

LR = LinearRegression()

model = LR.fit(X_train, y_train)
```

```python
# Checking the accuracy of the model

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
      
```

```python
y_predict = model.predict(X_test)

print(y_predict)
```

```python
plt.scatter(y_test, y_predict)
plt.xlabel('Test Value')
plt.ylabel('Predict Value')
plt.title('Actual vs Predicted')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_test, y_predict))
```
